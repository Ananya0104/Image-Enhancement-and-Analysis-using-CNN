#STRIDE

Effect of change of stride while performing convolution on images.

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
import cv2
import numpy as np
```

```python
# Load an example image
image_path = '/content/BIRD.jpeg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

image = cv2.resize(image, (1126, 980))

# Add batch dimension
image = np.expand_dims(image, axis=0)
image = np.expand_dims(image, axis=3)

# Define a simple convolutional neural network with a single convolutional layer and striding
model = models.Sequential()
model.add(layers.Conv2D(1, (3, 3), strides=(1, 1), activation='relu', input_shape=(980, 1126, 1)))

# Create a kernel/filter for demonstration purposes

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]], dtype=np.float32).reshape((3, 3, 1, 1))

# Set the model's weights to the custom kernel
model.layers[0].set_weights([kernel, np.array([0.0])])

# Apply convolution
result = model.predict(image)
```

```python
#Display original and convolved images
plt.figure(figsize=(8, 6))
plt.imshow(image[0, :, :, 0], cmap='gray')
plt.title('Original Image')
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(result[0, :, :, 0], cmap='gray')
plt.title('Convolved Image with Stride (1, 1)')

plt.show()
```

```python
model2 = models.Sequential()
model2.add(layers.Conv2D(1, (3, 3), strides=(2, 2), activation='relu', input_shape=(980, 1126, 1)))

# Set the model's weights to the custom kernel
model2.layers[0].set_weights([kernel, np.array([0.0])])

# Apply convolution
result2 = model2.predict(image)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(result2[0, :, :, 0], cmap='gray')
plt.title('Convolved Image with Stride (2, 2)')
plt.show()
```

```python
model3 = models.Sequential()
model3.add(layers.Conv2D(1, (3, 3), strides=(5, 5), activation='relu', input_shape=(980, 1126, 1)))

# Set the model's weights to the custom kernel
model3.layers[0].set_weights([kernel, np.array([0.0])])

# Apply convolution
result3 = model3.predict(image)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(result3[0, :, :, 0], cmap='gray')
plt.title('Convolved Image with Stride (5, 5)')
plt.show()
```
